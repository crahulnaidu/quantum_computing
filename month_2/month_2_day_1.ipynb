```python
from qiskit.circuit import QuantumCircuit,Parameter
```

```python
theta = Parameter("Оё")

qc = QuantumCircuit(3)
qc.rx(theta, 0)
qc.cx(0, 1)
qc.x(2)



qc.draw()
```

```python
from math import pi

angle_list = [pi / 3, pi / 2]
circuits = [qc.assign_parameters({theta: angle}) for angle in angle_list]

for circuit in circuits:
    display(circuit.draw())
```

```python
qc=QuantumCircuit(2)

theta1=Parameter("01")
theta2=Parameter("02")

qc.rx(theta1,0)
qc.rx(theta2,1)

qc.barrier()
qc.cx(0,1)

angles=[pi/2,pi/3,pi/4]

circuits=[qc.assign_parameters({theta1:angle,theta2:angle})for angle in angles]

for circuit in circuits:
  display(circuit.draw())
```
