```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator
```

```python
qc=QuantumCircuit(2)
qc.cz(0,1)

qc.draw()
```

```python
qc=QuantumCircuit(2)
qc.h(0)
qc.cx(0,1)
qc.cz(0,1)

state=Statevector.from_instruction(qc)
print(state)
```

```python
qc=QuantumCircuit(2)

qc.h([0,1])

#marking the state 01 by first converting into 11,applying cz and then converintg back to 01.
qc.x(0)
qc.cz(0,1)
qc.x(0)

state=Statevector.from_instruction(qc)
print(state)
```

```python
qc.h([0,1])
qc.measure_all()

print(AerSimulator().run(qc,shots=1000).result().get_counts())
#01 appears more often.
```

```python
## Explanation

#1. Created superposition of all inputs
#2. Used oracle to mark correct input via phase
#3. Used interference to amplify correct answer
#4. Measured to extract result
```
