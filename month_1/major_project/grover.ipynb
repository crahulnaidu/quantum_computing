```python
from qiskit import QuantumCircuit,transpile
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

```python
qc=QuantumCircuit(2,2)

#Create a superposition of all state's

qc.h([0,1])

#Orace,Mark the 11 state

qc.cz(0,1)

#Grover's diffusion operator

qc.h([0,1])
qc.x([0,1])
qc.h(1)
qc.cx(0,1)
qc.h(1)
qc.x([0,1])
qc.h([0,1])


qc.measure([0,1],[0,1])

aer=AerSimulator()

transpiled_circuit=transpile(qc,aer)

result=aer.run(transpiled_circuit).result()


print(result.get_counts())
```
