```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
```

```python
qc=QuantumCircuit(3,2)

qc.h([0,1,2])

#balanced oracle

qc.cx(0,2)
qc.cx(1,2)

qc.h([0,1])
qc.measure([0,1],[0,1])

qc.draw()
```

```python
result=AerSimulator().run(qc,shots=1000).result()

print(result.get_counts())
```
