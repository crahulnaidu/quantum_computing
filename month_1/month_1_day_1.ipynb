```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import numpy as np
```

```python
#A qubit is any superposition of states such that the probability for any state to appear when measured is |A|^2
```

```python
state=Statevector([1/np.sqrt(2),1/np.sqrt(2)])
qc=QuantumCircuit(1)
qc.initialize(state,0)
qc.measure_all()

result=AerSimulator().run(qc,shots=1000).result().get_counts()
print(result)
```

```python
def valid_qubit(alpha,beta):
  return abs(abs(alpha**2)+abs(beta**2)-1)<1e-6
```

```python
print(valid_qubit(1/np.sqrt(2),1/np.sqrt(3)))
```
