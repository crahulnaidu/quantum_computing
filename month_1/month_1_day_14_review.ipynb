```python
from qiskit import QuantumCircuit,transpile
from qiskit_aer import AerSimulator
```

```python
#grover's algorithm

def grover_circuit():
  qc=QuantumCircuit(2,2)

  qc.h([0,1])
  qc.cz(0,1)

  qc.h([0,1])
  qc.x([0,1])
  qc.h(1)
  qc.cx(0,1)
  qc.h(1)
  qc.x([0,1])
  qc.h([0,1])

  qc.measure([0,1],[0,1])
  return qc
```

```python
grover_circuit().draw()
```

```python
# Deutsch_jozsas algorithm

def deutsch_jozsa():
  qc=QuantumCircuit(3,2)
  qc.h([0,1,2])
  qc.cx(0,2)
  qc.cx(1,2)

  qc.h([0,1])
  qc.measure([0,1],[0,1])
  return qc
```

```python
deutsch_jozsa().draw()
```

```python
# Simon's algorithm

def simon():
  qc=QuantumCircuit(4,2)

  qc.h([0,1])

  qc.cx(0,2)
  qc.cx(1,2)
  qc.cx(0,3)
  qc.cx(1,3)

  qc.measure([2,3],[0,1])
  qc.h([0,1])
  qc.measure([0,1],[0,1])
  return qc
```

```python
simon().draw()
```

```python
sim = AerSimulator()

for func in [grover_circuit, deutsch_jozsa, simon]:
    qc = func()
    result = sim.run(qc, shots=1000).result()
    print(func.__name__, result.get_counts())
```

```python
"""Algorithm | Output Pattern | Key Mechanism
   Grover    | peaked         | amplitude amplification
   DJ        | deterministic  | interference
   Simon     | constraints    | linear algebra"""
```
