```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
```

```python
qc=QuantumCircuit(1)
qc.h(0)
qc.h(0)

state=Statevector.from_instruction(qc)
print(state)
```

```python
qc=QuantumCircuit(1,1)
qc.h(0)
qc.h(0)
qc.measure(0,0)
result=AerSimulator().run(qc,shots=1000).result().get_counts()
print(result)
```

```python
qc=QuantumCircuit(1)
qc.h(0)
qc.z(0)
qc.h(0)

state=Statevector.from_instruction(qc)
print(state)
```
